```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Step 1: Data (hours studied vs scores)
hours = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
scores = np.array([50, 55, 65, 70, 75, 85])

# Step 2: Create and train model
model = LinearRegression()
model.fit(hours, scores)

# Step 3: Predict
predicted = model.predict([[7]])
print("Predicted score for 7 hours:", predicted[0])

# Step 4: Plot graph
plt.scatter(hours, scores, label="Actual Data")
plt.plot(hours, model.predict(hours), label="Prediction Line")
plt.xlabel("Hours Studied")
plt.ylabel("Score")
plt.title("Student Score Predictor")
plt.legend()
plt.show()
```
